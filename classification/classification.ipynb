```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import sklearn 
import pandas as pd
import seaborn as sns
sns.set_theme()
```

### gathering the dataset for classification

```python
from sklearn.datasets import make_circles
x,y=make_circles(1000,noise=0.02,random_state=42)
```

```python
x[:5] , y[:5]
```

```python
circles = pd.DataFrame({'x1':x[:,0],
                       'x2':x[:,1],
                       'label':y})
```

```python
circles.head()
```

```python
sns.relplot(data=circles,x=circles.x1,y=circles.x2,hue=circles.label,palette=['blue','red'])
```

### split our data to train and test

```python
# changing numpy's float to float 32 (torch float)
x=torch.from_numpy(x).type(torch.float)
y=torch.from_numpy(y).type(torch.float)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```
