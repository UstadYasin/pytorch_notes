```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris=sns.load_dataset("iris")
iris.head()
```

ploting our data using seaborn

```python
sns.set_theme()
sns.pairplot(data=iris,hue="species",height=1.2)
```

```python
x_iris=iris.drop('species',axis=1)
y_iris=iris['species']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_iris,y_iris,random_state=42,test_size=0.2)

from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_model=model.predict(x_test)

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_model)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix 
cm=confusion_matrix(y_test,y_model)
print(cm)
```
