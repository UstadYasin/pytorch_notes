```python
a=torch.randint(1,100,(7,7))
a
```

```python
b=torch.randint(1,100,(1,7))
```

```python
print(torch.matmul(a,b))
```

```python
print(a.shape, b.shape)
# result as (7,1)
print(a @ b.t())

# or get a (7,) vector
print(a @ b.squeeze())

# alternative (uses broadcasting)
print((a * b).sum(dim=1))
```
