```python
import torch
from torch import nn
import matplotlib.pyplot as plt
```

```python
weight = 0.2
bias = 0.3
X = torch.arange(0,1,0.01).unsqueeze(dim=1)
y = bias*X + weight
```

```python
len(X),len(y)
```

```python
split=int(0.8*len(X))
split
```

```python
x_train , y_train = X[:split],y[:split]
```

```python
x_test , y_test = X[split:] , y[split:]
```

```python
def plt_predict(
    trainData=x_train,
    trainLabel=y_train,
    testData=x_test,
    testLabel=y_test,
    predictions=None
):
    plt.scatter(trainData,trainLabel,s=1,c='g',label='training data')
    plt.scatter(testData,testLabel,s=2,c="b",label="tasting data")
    if predictions is not None:
        plt.scatter(testData,predictions,c='r',s=4,label="predicted data")
    plt.legend()
```

```python
plt_predict()
```

```python
from torch import nn
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.weights=nn.Parameter(torch.randn(1,requires_grad=True,dtype=torch.float))
        self.bias=nn.Parameter(torch.randn(1,requires_grad=True ,dtype=torch.float))
    def forward(self,x : torch.Tensor)->torch.Tensor:
       return self.weights*x + self.bias
            
```

```python

m0=LinearRegressionModel()
list(m0.parameters())
```

```python
LinearRegressionModel()
```

```python
m0.state_dict()
```

```python
weight,bias
```

making prediction with our module 

```python
m0.eval()
with torch.inference_mode():
    y_p=m0(x_test)

y_p 
```

```python
plt_predict(predictions=y_p,)
```

```python
# setting up los function and optimizer
los_f=nn.L1Loss()

optimizer=torch.optim.SGD(params=m0.parameters(),lr=0.01)
```

```python
list(m0.parameters())
```

```python
torch.manual_seed(42)
epoch=100
for i in range(epoch):
    m0.train()
    y_pred=m0(x_train)
    loss=los_f(y_pred,y_train)
    # print("loss : ",loss)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    m0.eval()
    with torch.inference_mode():
        test_pred=m0(x_test)
        test_loss=los_f(test_pred,y_test)
    if epoch % 10 ==0:
        print(f"'epoch :'{epoch},'loss : '{loss} ,'test loss' {test_loss}")
```
